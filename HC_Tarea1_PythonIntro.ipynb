```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
from scipy.constants import g


#Declaramos la funcin para calcular la altura mxima/We declare the function
#to calculate the maximum height.
def max_height(data: dict):
   v0 = data["v0"]
   theta0 = data["theta0"]
   maxh = np.power((v0*(np.sin(theta0))),2)/(2*g)
   return maxh

#Creamos el arreglo de ngulos utilizando numpy/We create the array of angles using numpy.
mintheta = 0
maxtheta = m.pi/2
dtheta = (maxtheta - mintheta)/100
thetas = np.arange(mintheta, maxtheta, dtheta)

#Utilizamos las condiciones iniciales y la funcin para
#generar los datos/We use the initial conditions and the function to generate the data.
initialConditions = {"v0": 2.34, "theta0":thetas}
maxheights = max_height(initialConditions)


#Generamos la grfica/We generate the graph
fig, ax = plt.subplots(figsize=(10, 8))
ax.plot(thetas, maxheights)

fig.tight_layout()
ax.grid(True)
ax.minorticks_on()
ax.set_xlabel('ngulo inicial [Rad]')
ax.set_ylabel('Altura mxima [m]')
ax.set_title('Altura mxima vs ngulo inicial')
plt.savefig('HCGraficaTarea1.pdf', format="pdf", bbox_inches='tight', pad_inches=0.8)

plt.show()
```
